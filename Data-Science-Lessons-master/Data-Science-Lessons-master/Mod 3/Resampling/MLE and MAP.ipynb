# <center>Maximum Likelihood Estimation

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [2, 6, 3, 4, 4, 5, 5, 5, 8, 9]
z = [0,0,0,0,0.05,0,0,0.05,0,0]
```

```python
plt.scatter(data,z,c='r',s=200)
plt.ylim(0,0.1)
```

```python
mu = 4
sigma = 1
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.scatter(data,z,c='r',s=200)
```

```python
sum(stats.norm(mu,sigma).pdf(z))
```

```python
mu = np.mean(data)
sigma = np.std(data)
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.scatter(data,z,c='r',s=200)
```

```python
sum(stats.norm(mu,sigma).pdf(z))
```

# <center> MLE vs MAP

## <center> MAP - Maximum A Posteriori Estimation

Consider flipping a coin 10 times...

```python
flips = ['H', 'T', 'H', 'H', 'H', 'H', 'T', 'T', 'H','T']
```

What are the odds of heads?

MLE approach...

```python
flips.count('H'), flips.count('T')
```

```python
odds_of_heads = flips.count('H')/len(flips)
print('The odds of heads is:', odds_of_heads)
```

MAP approach...

```python
prior_prob = 0.5
odds_of_heads = flips.count('H')/len(flips)
```

```python
(prior_prob+odds_of_heads)/2
```

### MAP combines prior knowledge and the observed knowledge 

Say we roll a die 10 times and get the following results

```python
rolls = [1,5,2,2,5,6,4,6,1,5]
```

```python
## MLE
probs = [rolls.count(x)/len(rolls) for x in range(1,7)]
probs
```

```python
## MAP
prior_probs = [1/6]*6
map_probs = [(x+y)/2 for x,y in zip(probs,prior_probs)]
map_probs
```
